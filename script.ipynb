# Creating Linear Layer

```python
input_tensor = torch.tensor([[0.3471, 0.4547, -0.2356]])

# Create a Linear layer: 3 inputs -> 2 outputs
#output = input @ W^T + B
#W - weights
#B - bias
linear_layer = nn.Linear(
                         in_features=3, #Inputs
                         out_features=2 #number of neurons
                        )

# Pass input_tensor through the linear layer
output = linear_layer(input_tensor)


print(f"Wagi:  {linear_layer.weight}") #matrix[2x3]
print(f"Bias: {linear_layer.bias}") #vector[2]
print(f"Output: {output}") #output [1x2]
```

# Neural Network with 2 layers (Sequential connects layers)

```python
input_tensor = torch.Tensor([[2, 3, 6, 7, 9, 3, 2, 1]])

#model: 8 -> 4 -> 1
model = nn.Sequential(nn.Linear(8,4), # Layer 1: 8 inputs -> 4 neurons
                      nn.Linear(4,1)  # Layer 2: 4 inputs -> 1 neuron(output)
                    )

# First Layer Parameters (8 -> 4)
print(f"Wagi0: {model[0].weight}") # [4x8] - 4 neurons x 8 weights each
print(f"Bias0: {model[0].bias}")   # [4] - 1 bias on each neuron

#Wagi i bias dla 2 neurona
print(f"Wagi1: {model[1].weight}") # [1x4] - 1 neuron x 4 weights
print(f"Bias1: {model[1].bias}")   # [1] - 1 bias

# Flow: [8] -> Layer 1 -> [4] -> Layer 2 -> [1]
output = model(input_tensor)
print(f"Output: {output}")
```

# Activation Function Sigmoid
Converts any number to value between 0 and 1 (like probability) 1/(1 + e^-x)<br>
It is for binary classification

```python
input_tensor = torch.tensor([[2.4]])

sigmoid = nn.Sigmoid()

probability = sigmoid(input_tensor)
print(probability)
```

# Activation Function Softmax

Converts a vector of numbers into probability distribution<br>
All outputs sum to 1.0 (like class probabilities)

```python
input_tensor = torch.tensor([[1.0, -6.0, 2.5, -0.3, 1.2, 0.8]])

# Create a softmax function and apply it on input_tensor
softmax = nn.Softmax()
probabilities = softmax(input_tensor)
print(probabilities)
```

# Small Neural Network for binary classification

```python
input_tensor = torch.Tensor([[3, 4, 6, 2, 3, 6, 8, 9]])

model = nn.Sequential(
  nn.Linear(8,1),
  nn.Sigmoid()
)

output = model(input_tensor)
print(output)
```

# Network for multi class classification

```python
input_tensor = torch.Tensor([[3, 4, 6, 7, 10, 12, 2, 3, 6, 8, 9]])

# 4 Layers
model = nn.Sequential(
  nn.Linear(11, 20),
  nn.Linear(20, 12),
  nn.Linear(12, 6),
  nn.Linear(6, 4),
  nn.Softmax()
)

output = model(input_tensor)
print(output)
```

# One-hot encoding <br>
converts a class label (integer) into a binary vector

```python
y = 1 # class label (0, 1, 2)
num_classes = 3 # total number of classes

# one-hot encoded vector using NumPy
one_hot_numpy = np.array([0, y, 0])

# one-hot encoded vector using PyTorch
one_hot_pytorch = F.one_hot(torch.tensor(y), num_classes=3)

print("One-hot vector using NumPy:", one_hot_numpy)
print("One-hot vector using PyTorch:", one_hot_pytorch)
```
